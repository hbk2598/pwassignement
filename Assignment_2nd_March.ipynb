```python
#Solution2
'''A scatter plot identifies a possible relationship between changes observed in two different sets of variables.
It provides a visual and statistical means to test the strength of a relationship between two variables.'''
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
```

```python
x
```

```python
y
```

```python
import matplotlib.pyplot as plt
plt.scatter(x,y)
plt.xlabel(" X Axis")
plt.ylabel("Y Axis")
plt.title(" X vs Y")
```

```python
#Solution3
'''subplot() function adds subplot to a current figure at the specified grid position.  
It is similar to the subplots() function however unlike subplots() it adds one subplot at a time.'''
```

```python
import numpy as np
x = np.array([0, 1, 2, 3, 4, 5]) 
y = np.array([0, 100, 200, 300, 400, 500])
```

```python
x
```

```python
y
```

```python
import matplotlib.pyplot as plt
plt.subplot(121)
plt.plot(x, y, marker="x", color="green")
```

```python
#Solution4
'''Bar plots are a type of data visualization used to represent data in the form of rectangular bars. The height of each bar represents the value of a data point,
and the width of each bar represents the category of the data.'''
```

```python
import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
```

```python
import matplotlib.pyplot as plt
plt.bar(company,profit)
```

```python
plt.barh(company,profit,color='red')
```

```python
#Solution5
'''A Box Plot is also known as Whisker plot is created to display the summary of the set of data values having properties like minimum,
first quartile, median, third quartile and maximum'''
```

```python
import numpy as np
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
box=[box1,box2]
```

```python
import matplotlib.pyplot as plt
plt.boxplot(box)
```
