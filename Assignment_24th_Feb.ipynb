```python
#Solution9
import pandas as pd

def solution9(df):
    mean=df['Values'].mean()
    median=df["Values"].median()
    std=df["Values"].std()
    
    print('Mean:', mean)
    print('Medain:', median)
    print('Standard deviation:', std)
```

```python
data={'Values':[25,30,27,30,40]}
df=pd.DataFrame(data)
```

```python
solution9(df)
```

```python
#Solution10
def solution10(df,Column_name,window_size):
    moving_average=df[Column_name].rolling(window_size).mean()
    return moving_average
```

```python
import pandas as pd
df=pd.DataFrame({'Sales':[1,2,3,4,5,6,7,8],'Date':pd.to_datetime(['03-11-2023','04-11-2023','05-11-2023','06-11-2023','07-11-2023','08-11-2023','09-11-2023','10-11-2023'])})
```

```python
moving_average=solution10(df,'Sales',7)
df["MovingAverage"]=moving_average
print(df)
```

```python
#Solution11
import pandas as pd
df=pd.DataFrame({'Date':['2023-11-03','2023-11-04','2023-11-05','2023-11-06','2023-11-07','2023-11-08','2023-11-09']})
```

```python
def solution11(df):
    df['Date']=pd.to_datetime(df['Date'])
    df['Weekday']=df['Date'].dt.day_name()
    return df
```

```python
solution11(df)
```

```python
#Solution12
def solution12(df,start_date,end_date):
    df_filtered=df[df['Date'].between(start_date,end_date)]
    return df_filtered
```

```python
import pandas as pd
df=pd.DataFrame({'Date':['2023-11-03','2023-10-04','2023-01-05','2023-01-06','2023-11-07','2023-01-28','2023-11-09']})
```

```python
df_filtered=solution12(df,'2023-01-01','2023-01-31')
print(df_filtered)
```
