```python
#Question3
'''ArithmeticError is simply an error that occurs during numeric calculations.
ArithmeticError class have three errors
(i)OverFlowError
(ii)ZeroDivisionError
(iii)FloatingPointError'''
```

```python
#OverflowError
j = 5.0

try:
    for i in range(1, 1000):
        j = j**i
except ArithmeticError as e:
    print(f"{e}, {e.__class__}")
```

```python
#ZeroDivisionError
try:
    1/0
except ArithmeticError as e:
    print(f"{e}, {e.__class__}")
```
